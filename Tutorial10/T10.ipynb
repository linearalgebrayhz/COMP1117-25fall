## Python String Review

A `String` in Python is a sequence of characters. You can create a string by enclosing characters in single quotes (`'`) or double quotes (`"`).

Triple quotes (`'''` or `"""`) can be used for multi-line strings.

String is immutable, meaning once created, its content cannot be changed.

```python
string_2 = "123123 \n456"

string_3 = """123123 
456
"""

print(string_2)
print(string_3)
```

```python
s = "COMP1117 Tutorial10"
s[2] = "m"
```

```python
a = "COMP1117"
print(f"The id of string a: {id(a)}")
b = " Tutorial"
print(f"The id of string b: {id(b)}")

a += b 

print(f"The id of string a+b: {id(a)}")

print(f"The id of string b*2: {id(b*2)}")
```

```python
# An example of Python OOP, you can play around with it
import math

class complexNum:

    def __init__(self, re, im) -> None: # This is called constructor method
        # initialize
        self.re = re
        self.im = im

    def __add__(self, c): # This is called operator overloading
        return complexNum(self.re + c.re, self.im + c.re)
    
    def __neg__(self):
        return complexNum(-self.re, self.im)

    def __repr__(self) -> str: # this method determines what the object look like 
        return f"{self.re} + {self.im}i"
    
    @staticmethod # This is a method bind to the whole complex class
    def modulus(c):
        return math.sqrt(c.re ** 2 + c.im ** 2)

    
c1 = complexNum(1,2)
print(c1)
c2 = complexNum(1,1)
print(c1+c2) 

print(complexNum.modulus(c1))
```
