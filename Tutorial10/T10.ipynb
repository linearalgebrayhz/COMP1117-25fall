```python
# An example of Python OOP
import math

class complexNum:

    def __init__(self, re, im) -> None:
        self.re = re
        self.im = im
        pass

    def __add__(self, c):
        return complexNum(self.re + c.re, self.im + c.re)
    
    def __neg__(self):
        return complexNum(-self.re, self.im)

    def __repr__(self) -> str:
        return f"{self.re} + {self.im}i"
    
c1 = complexNum(1,2)
print(c1)
```
