## 2. Tutorial Exercises

```python
def f(n): # f(n)
    if n == 0:
        return 1
    if n % 2:
        n -= 1
    return n * f(n-2)

n = int(input())
print(f(n))
```

```python
def multiply(a, b):
    if b == 0:
        return 0
    return a + multiply(a, b-1)

a = int(input())
b = int(input())

print(multiply(a,b))
```

```python
class complex:

    def __init__(self,re,im) -> None:
        self.re = re
        self.im = im

    def __repr__(self) -> str:
        return f"{self.re} + {self.im}i"
    
num = complex(1,2)
print(num)
```
